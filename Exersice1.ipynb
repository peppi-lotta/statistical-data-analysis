```python
data = (170, 192, 184, 165, 176, 181, 163)
data2 = (170, 170, 170, 170, 192, 192, 192, 192, 184, 184, 184, 184, 165, 165, 165, 165, 176, 176,
176, 176, 181, 181, 181, 181, 163, 163, 163, 163)

# Shapiro-Wilk test is a test of normality. 
# Scipy.stats has a ready made fuction for this so we are using it. 

print("3")
shap = shapiro(data).pvalue
shap2 = shapiro(data2).pvalue
print("Pvalue from Shapiro-Wilk test:", shap)
print("Pvalue from Shapiro-Wilk test:", shap2)

# If the pvalue is below a certain significant level, 
# then we have sufficient evidence to say the sample data does not come from a normal distribution
# Pvalue needs to be more than 0.05 to say the data is from normally distributed dataset 
# (P < 0.05 -> normally distributed data)
#         -> based on this: shap is normally distributed and shap2 is not 
# These might be a part of a larger dataset that is part of normal distribution, 
# 

# Kolmogorov-Smirnov test also in scipy. Another way of checking for normal distribution.
# Better (than Shapiro-Wilk) for sets with more than 50 data points. 

# Pvalue: Propability of Null hapothesis (distribution is normal). Propability of the dataset 
# being a part of larger dataset that is normally distributed. Pvalue is only an indicator.
# The larger dataset can be normally distributed even if the subset is not. 
```
