```python
from scipy.stats import shapiro

data = (170, 192, 184, 165, 176, 181, 163)
data2 = (170, 170, 170, 170, 192, 192, 192, 192, 184, 184, 184, 184, 165, 165, 165, 165, 176, 176,
176, 176, 181, 181, 181, 181, 163, 163, 163, 163)

# Shapiro-Wilk test is a test of normality. Scipy.stats has a ready made fuction for this so we are using it. 

print("3")
shap = shapiro(data)
shap2 = shapiro(data2)
print(shap)
print(shap2)

# If the pvalue is below a certain significant level, then we have sufficient evidence to say the sample data does not come from a normal distribution
# Pvalue needs to be more than 0.05 to say the data is from normally distributed dataset (P < 0.05 -> normally distributed data)
#         -> based on this: shap is normally distributed and shap2 is not 
```

```python
import pandas as pd

print("4")
data = (170, 192, 184, 165, 176, 181, 163)
data2 = (170, 170, 170, 170, 192, 192, 192, 192, 184, 184, 184, 184, 165, 165, 165, 165, 176, 176,
176, 176, 181, 181, 181, 181, 163, 163, 163, 163)
bns = [150, 160, 170, 180, 190, 200]

print("First data set")
print(data)
pd.DataFrame(data).hist(bins=bns, color="pink")
pd.DataFrame(data).plot(kind='density', color="pink")

print("Second data set")
print(data2)
pd.DataFrame(data2).hist(bins=bns, color="pink")
pd.DataFrame(data2).plot(kind='density', color="pink")
```

```python
import pandas as pd  

print("5")
# data location 
loc = r'C:\Users\peppi\Desktop\lipasto\statisticalDataAnalysis\graph.txt'

#import data 
mydata = pd.read_table(loc, sep="\s+")

#DataFrame mydata
df = pd.DataFrame(mydata)
print(df)

all_values = df['No_Smoothing']
print(all_values)

#valus from year 2000 and up, dropped NaN values and reset index
selection = df['No_Smoothing'].where(df['Year'] > 1999).dropna().reset_index(drop=True)
#selection = df.loc[df["Year"] >= 2000]["No_Smoothing"]
print(selection)

#mean
print("Mean of all values:", pd.DataFrame(all_values).mean())
print("Mean of values from year 2000 and up:", pd.DataFrame(selection).mean())

#median
print("Median of all values:", pd.DataFrame(all_values).median())
print("Median of values from year 2000 and up:", pd.DataFrame(selection).median())

#density plots here
pd.DataFrame(all_values).plot(kind='density', color="pink")
pd.DataFrame(selection).plot(kind='density', color="pink")

```
