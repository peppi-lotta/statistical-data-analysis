<h3>Invalid values</h3>
Distance column has nagative values. Otherwise data seems complete and intact. Distance should never be negative in this situation. Here we determine the best action for these negative values.

```python
# display the rows with the negative distances
print("Negative:")
dist_neg_data = original_data.loc[original_data['distance'] < 0 ]
display(dist_neg_data)

# original_data without the rows with negative distances
dist_pos_data = original_data.loc[original_data['distance'] >= 0 ].reset_index()

print("dist_pos_data normality check:", stats.shapiro(dist_pos_data['distance']))
# check if negative values turned to positive are likely to come from the same dataset as positive values
# Unpaired and not normal -> Mann-Whitney U test
neg_list = dist_neg_data['distance'].copy()
pos_list = []
for i in neg_list:
   pos_list.append(i * (-1))
print(pos_list)
print(stats.mannwhitneyu(dist_pos_data['distance'], pos_list))

data = dist_pos_data.copy()
data
```

There are 11 rows where the distance traveled is negative. The rows seem have all the other data correctly displayed. I did a check with Mann_whitney U test to check if negative values turned to positive are likely to come from the same dataset as the other already positive values alone. P-value was 0.042. Statistical evidence shows that the values are not like from the same data set. Based on this I will exclude the rows with negative values.

<h2>Hypothesis testing</h2>

<h3>Explain what the p-value of a statistical test means and how it can be used to make conclusions.</h3>

P-value indicates how likely a null hypothesis is. Smaller the p-value is higher the change that the null hypothesis is not true. 
If p-value is smaller than 0.05 null hypothesis should be rejected.

<h3>Is there statistical evidence to claim that the travel times tend to be shorter or longer for the single than for the season ticket type?</h3>

```python
#single
single_dur = single[single['duration'] >= 0]['duration']
print("Single:", stats.shapiro(single_dur))
plt.hist(single_dur, bins = 50, color='pink')
plt.show()

#season
season_dur = season['duration']
print("Season:", stats.shapiro(season_dur))
plt.hist(season_dur, bins = 50, color='pink')
plt.show()

# these are not normally distributed and unpaired
# so we use Mann-Whitney U test
test = stats.mannwhitneyu(single_dur, season_dur)
print(test)

# means
print("Single:", "Mean:", single_dur.mean(), "Median:", single_dur.median())
print("Season:", season_dur.mean(), "Median:", season_dur.median())
```

Mann-Whitney U's null hypothesis is that set come from same distribution. P-value is clearly over 0.05 so this indicates statistical evidence that single and savonia ticket users don't have significant difference regarding duration. Values could be from the same distribution. 

<h3>Is there statistical evidence to claim that the single and savonia ticket types differ with respect to how often the electric assistance is used?</h3>

```python
# comparing categorial values -> Chi-Pearson squared test
# concat data into one set that can be crosstabed
chi_data = pd.concat([single['assistance'],savonia['assistance']], keys=['single', 'savonia']).reset_index()
display(chi_data)

# crosstab the chi_data
cros = pd.crosstab(index=chi_data['assistance'], columns=chi_data['level_0'])
display(cros)

#Chi-Pearson square test
chi = stats.chi2_contingency(cros)
print("p =", chi[1])
```

Based on the p-value single and savonia clearly differ on how often the assistance is used. Chi-Pearson square test was used to get the P-value as both (ticket and assitance) are categorial. 
<br>
Users that have savonia tickets use assistance <strong>95.9%</strong> of the time. Where as single ticket users use it only <strong>21.5%</strong> of the time. 

<h3>Is there statistical evidence to claim that the travel distances tend to be shorter or longer during one month than during the others?</h3>

```python
# month selection
april = data.loc[data['month'] == 4].reset_index()
may = data.loc[data['month'] == 5].reset_index()
june = data.loc[data['month'] == 6].reset_index()
july = data.loc[data['month'] == 7].reset_index()
august = data.loc[data['month'] == 8].reset_index()
september = data.loc[data['month'] == 9].reset_index()
october = data.loc[data['month'] == 10].reset_index()

#make distance dataframe with months as columns
d = {'april': april.distance, 
    'may': may.distance, 
    'june': june.distance, 
    'july': july.distance, 
    'august': august.distance,
    'september': september.distance,
    'october': october.distance}
months = pd.DataFrame(data=d)
display(months)

# Description
print("Description:")
months.describe()
```

```python
# Multiple non-normally distributed variables
# -> Kruskal-Wallis test

# all
kruskal = stats.kruskal(april.distance, may.distance, june.distance, july.distance, august.distance, september.distance, october.distance).pvalue
print("All:", kruskal)

# based on description months chosen that have similar means and standad diviations
kruskal = stats.kruskal(april.distance, june.distance, july.distance, august.distance).pvalue
print("April/June/July/August:", kruskal)

# testing if may fits in 
kruskal = stats.kruskal(may.distance, april.distance, june.distance, july.distance, august.distance).pvalue
print("MAY/April/June/July/August:", kruskal)

# testing if september fits in 
kruskal = stats.kruskal(september.distance, april.distance, june.distance, july.distance, august.distance).pvalue
print("SEPTEMBER/April/June/July/August:", kruskal)

# testing if october fits in 
kruskal = stats.kruskal(october.distance, april.distance, june.distance, july.distance, august.distance).pvalue
print("OCTOBER/April/June/July/August:", kruskal)
```

Kruskal-Wallis of all months tells us that there is statistical evidence to suggest a difference in distances between the months. Based on description april, june, july and, august have similar standard diviation and means. Kruskal-Wallis confirms this with p-value of 0.51. <br>

I add may, september and october alternately to the Kruskal-Wallis with the months already establish as not having notable difference. This shows that all these months have a notable difference. October's mean is the same as the compared months but standard deviation is is smaller. Distances are not notably longer or shorter. <br>

May has shortest travel distances and september the longest based on the means.
