<h2>3.</h2>
Download and unzip the DataSet SOFT file from <a href="https://www.ncbi.nlm.nih.gov/sites/GDSbrowser?acc=GDS5037">https://www.ncbi.nlm.nih.gov/sites/GDSbrowser?acc=GDS5037</a>. The SOFT file format contains tabular gene expression data accompanied by its metadata. The rows that start with either ^, !, or # are metadata, which can be ignored in this exercise.<br><br>

Some useful pieces of information about the file GDS5037.soft:<br>

The first 160 rows and the last row are metadata.<br>
The ID_REF column contains probe ids. Use them as row names.<br>
The IDENTIFIER column contains gene names.<br>
The other columns contain gene expression levels for subjects.<br>
Load and filter the data such that your data frame only contains the probe and subject ids listed in the files 43-probes.ids and 43-subjects.ids, respectively.<br>

* For each subject, calculate the mean expression level over the genes. Do you notice any obvious differences between the subjects?

* For each probe, calculate the mean expression level over the subjects. Do you notice any obvious differences between the probes?

* How many probes are there with a mean greater than 10.0? Which genes do they correspond to?

```python
# metadata = list(range(160)) + [41269] and instead of header and skipfooter -> skiprows=metadata
soft = pd.read_csv('GDS5037.soft', sep="\s+" ,header=160, skipfooter=1, engine='python')
df = pd.DataFrame(soft)
df_full = df.set_index('ID_REF')
df_full
```

```python
probes = pd.read_table('43-probes.ids', header=None)[0]
subjects = pd.read_table('43-subjects.ids', header=None)[0]
df = df_full.loc[probes,subjects]
df_full
```

```python
#subject mean
df.mean()
```

```python
#probe means
df.mean(axis=1)
```

```python
# How many probes are there with a mean greater than 10.0? Which genes do they correspond to?
df = df.loc[df.mean(axis=1) > 10]
ids = df.index.values
identefiers = df_full.loc[ids].IDENTIFIER
identefiers
```

```python
# Import helsinki and espoo data
helsinki = pd.DataFrame(pd.read_csv('44-helsinki.csv', sep="," ,header=0, index_col=0))
espoo = pd.DataFrame(pd.read_csv('44-espoo.csv', sep="," ,header=0, index_col=0))
print(helsinki.shape, espoo.shape)

#right join data on date
rm = pd.merge(espoo, helsinki, how="right", on=["date"])
# rm = pd.concat([helsinki, espoo], axis=1)
rm
```

```python
# Total observations 3134

# Observations for each street 
rm.count()
```

```python
df = rm.dropna();
df = df.reset_index(drop=True)
df
# Observations with all 1400
```

```python
col_sum = rm.sum().sort_values(ascending=False)
col_sum
# Merikannontie was busiest
```
