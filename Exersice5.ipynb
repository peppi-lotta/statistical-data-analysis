1. Create a box plot of gene expression levels for each subject. Do you notice visible differences between the subjects or between the groups?

4. How many differentially expressed genes (i.e. statistically significant differences) are there at the false discovery rate of 0.05?

```python
# 2.1
soft = pd.read_csv('GDS5037.soft', sep="\s+" ,header=160, skipfooter=1, engine='python')
df_soft = pd.DataFrame(soft)
df_soft = df_soft.set_index('ID_REF')
```

```python
# choose based on files
probes = pd.read_table('52-probes.ids', header=None)[0]
controls = pd.read_table('52-control.ids', header=None)[0]
asthma = pd.read_table('52-asthma.ids', header=None)[0]
subjects = pd.concat([controls, asthma])

df_full = df_soft.loc[probes,subjects]
df_full
```

```python
# Use a T-test to find differentially expressed genes (i.e. probes for which the means are different between the control and asthma groups). 
# Adjust the p-values with the Benjamini-Hochberg method.
df_control = df_full.loc[:,controls]
df_asthma = df_full.loc[:,asthma]

dft = pd.DataFrame(columns=['result_of_t-test'], index=[df_control.index])
for ind in df_control.index:
    dft.at[ind, 'result_of_t-test'] = stats.ttest_ind( df_control.loc[ind, :], df_asthma.loc[ind, :] ).pvalue
    
newdf = dft.sort_values(by='result_of_t-test')
ids = newdf.head(10).index.values
for id in ids:
    print(df_soft.loc[id].IDENTIFIER)
```
