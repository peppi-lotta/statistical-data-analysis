```python
import pandas as pd 
import scipy
from scipy import stats
import numpy as np 
```

- Test two normally distributed sets: Unpaired and paired t-test
- Test two unpaired and not normally distributed variables: Mann–Whitney U (Use only when the number of observation in each sample is n > 20)
- Test two paired and not normally distributed variables: Wilcoxon signed-rank test
- Test categorial values: Pearson's Chi-squared test (frequencies should be at least 5)

```python
print("1")

dataset_a = pd.Series([34, 23, 52, 47, 34])
dataset_b = pd.Series([48, 28, 33, 46, 41, 35])
dataset_c = pd.Series([34, 53, 51, 35, 52, 32])

#S-W test for datasets to determine normality (P > 0.05 = normal)
shapiro_test_a = stats.shapiro(dataset_a)
shapiro_test_b = stats.shapiro(dataset_b)
shapiro_test_c = stats.shapiro(dataset_c)
# Sets A and B are normally distributed C is not  

print('Pvalue for dataset A according to Shapiro-Wilks test:', shapiro_test_a.pvalue)
print('Pvalue for dataset B according to Shapiro-Wilks test:', shapiro_test_b.pvalue)
print('Pvalue for dataset C according to Shapiro-Wilks test:', shapiro_test_c.pvalue, '\n')

# Compare A&B
# Unpared t-test
# These can't be related as they don't have the same amount of datapoints
compare_a_b = stats.ttest_ind(dataset_a, dataset_b)
print('Pvalue for dataset A and B according to unpaired t-test:', compare_a_b.pvalue, '\n')

# Compare A&C 
# Mann–Whitney U test
compare_a_c = stats.mannwhitneyu(dataset_a,dataset_c,use_continuity=True,alternative='two-sided')
print('Pvalue for dataset A and C according to Mann–Whitney U test:', compare_a_c.pvalue, '\n')

# Compare B&C 
# Mann–Whitney U test and Wilcoxen both as we don't know id the data is paired 
compare_b_c = stats.mannwhitneyu(dataset_b,dataset_c,use_continuity=True,alternative='two-sided')
print('Pvalue for dataset B and C according to Mann–Whitney U test:', compare_b_c.pvalue)

wilcox_b_c = scipy.stats.wilcoxon(dataset_b,dataset_c)
print('Pvalue for dataset B and C according to Wilcoxon signed-rank test', wilcox_b_c.pvalue, '\n')
# No significat difference ether way
```

```python
print("3")

# test and retest of same group are paired 
# values between groups are not paired 

d_test = pd.Series([5.6, 3.1, 8.7, 4.5, 6.7, 4.5])
d_retest = pd.Series([6.1, 5.8, 8.5, 5.3, 7.2, 5.1])

e_test = pd.Series([4.5, 3.9, 7.1, 4.3, 6.9, 8.2, 7.6])
e_retest = pd.Series([4.9, 4.7, 7.8, 4.8, 7.5, 7.8, 8.1])

shapiro_dt = stats.shapiro(d_test)
shapiro_drt = stats.shapiro(d_retest)
print('Pvalue for D test dataset according to Shapiro-Wilks test:', shapiro_dt.pvalue)
print('Pvalue for D retest dataset according to Shapiro-Wilks test:', shapiro_drt.pvalue, "\n")

shapiro_et = stats.shapiro(e_test)
shapiro_ert = stats.shapiro(e_retest)
print('Pvalue for E test dataset according to Shapiro-Wilks test:', shapiro_et.pvalue)
print('Pvalue for E retest dataset according to Shapiro-Wilks test:', shapiro_ert.pvalue)

# D test and retest are normally distributed
# E test is normally distributed and retest is not

# D test/retest compare (paired, both normally distributed)
# Paired t-test
print("a)")
ttest_rel = stats.ttest_rel(d_test, d_retest)
print("Pvalue from Paired t-test:", ttest_rel.pvalue)

# E test/retest compare (paired, both NOT normally distributed)
# Wilcoxon signed-rank test
print("b)")
wilcoxon = scipy.stats.wilcoxon(e_test,e_retest)
print("Pvalue from Wilcoxon signed-rank test:", wilcoxon.pvalue)

# D test / E test compare (NOT paired, both normally distributed)
# Unpared t-test
print("c)")
ttest_ind = stats.ttest_ind(d_test, e_test)
print("Pvalue from Unaired t-test:", ttest_ind.pvalue)

# D retest / E retest compare (NOT paired, both NOT normally distributed)
# Mann–Whitney U test
print("d)")
mwu = stats.mannwhitneyu(d_retest,e_retest,use_continuity=True,alternative='two-sided')
print("Pvalue from Mann–Whitney U test:", mwu.pvalue)
```

```python
print("4")

# Pearson correlation and related P-value.
print("a)")
pearson = scipy.stats.pearsonr(e_test,e_retest)
print("r:", pearson.statistic, " p:", pearson.pvalue)

# Spearman correlation and related P-value.
print("b)")
spearman = scipy.stats.spearmanr(e_test,e_retest)
print("r:", spearman.correlation, " p:", spearman.pvalue)

# Rank the two groups. (For same values use half. 
# E.g. if there are two values with the smallest value, then rank them both 1.5.)
print("c)")
print("E test data:")
et_rank = e_test.rank()
et  = pd.DataFrame({'Value' : e_test, 'Rank' : et_rank})
print(et, "\n")

print("E retest data")
ert_rank = e_retest.rank()
ert  = pd.DataFrame({'Value' : e_retest, 'Rank' : ert_rank})
print(ert,  "\n")

# Pearson correlation and related P-value of the ranking obtained in 3).
print("d)")
pearson = scipy.stats.pearsonr(et_rank,ert_rank)
print("r:", pearson.statistic, " p:", pearson.pvalue)
```
